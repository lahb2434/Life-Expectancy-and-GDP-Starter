```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('all_data.csv')
df = df.rename(columns={'Life expectancy at birth (years)':'LE'})

sns.set(style="whitegrid")
ax = sns.scatterplot(
    y="GDP",
    x="LE",
    hue="Country",
    data=df,
    palette="Set2",
    legend=True,
    linewidths=0,
)

# Add a title and labels
ax.set_title("GPD vs. Life Expectancy")
ax.set_ylabel("GDP per capita")
ax.set_xlabel("Life expectancy")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for demonstration
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]  # First set of y-values (smaller scale)
y2 = [1000000, 2000000, 1500000, 1800000, 1200000]  # Second set of y-values (larger scale)

fig, ax1 = plt.subplots(figsize=(8, 6))

# Plotting the first line on the primary y-axis
ax1.plot(x, y1, label='Line 1', color='blue')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis (Line 1)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis and plot the second line
ax2 = ax1.twinx()  # Create a twin axes sharing the x-axis
ax2.plot(x, y2, label='Line 2', color='red')
ax2.set_ylabel('Y-axis (Line 2)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Two Lines with Different Scales')
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your DataFrame)
data = {
    'Year': [2000, 2001, 2002, 2003, 2004],
    'Country': ['Country1', 'Country1', 'Country1', 'Country1', 'Country1'],
    'Life_Expectancy': [75, 76, 77, 78, 79],  # Life expectancy values for Country1
    'GDP': [1000, 1200, 1500, 1800, 2000]  # GDP values for Country1
    # Add life expectancy and GDP values for other countries (Country2, Country3, ...)
}

df = pd.DataFrame(data)

# Create separate subplots for each country's Life Expectancy and GDP
countries = df['Country'].unique()

for country in countries:
    fig, ax1 = plt.subplots(figsize=(8, 4))
    
    data_country = df[df['Country'] == country]
    year = data_country['Year']
    
    ax1.plot(year, data_country['Life_Expectancy'], label='Life Expectancy', color='blue')
    ax1.set_ylabel('Life Expectancy', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    
    ax2 = ax1.twinx()
    ax2.plot(year, data_country['GDP'], label='GDP', color='red')
    ax2.set_ylabel('GDP', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    
    plt.title(country)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Country': ['Chile', 'Chile', 'Chile', 'Chile', 'Chile','Turkey', 'Turkey', 'Turkey', 'Turkey', 'Turkey'],
    'Year': [2000, 2001, 2002, 2003, 2004, 2000, 2001, 2002, 2003, 2004],
    'LE': [77.3, 77.3, 77.8, 77.9, 78.0, 77.3, 77.3, 77.8, 77.9, 78.0],
    'GDP': [7.786093e+10, 7.097992e+10, 6.973681e+10, 7.564346e+10, 9.921039e+10, 7.786093e+10, 7.097992e+10, 6.973681e+10, 7.564346e+10, 9.921039e+10],
}

df = pd.DataFrame(data)

# Get unique country names
countries = df['Country'].unique()

# Create separate plots for each country
num_plots = len(countries)
rows = 3
cols = 2
fig, axes = plt.subplots(rows, cols, figsize=(12, 10))

for i, country in enumerate(countries):
    row = i // cols
    col = i % cols
    ax = axes[row, col]
    country_data = df[df['Country'] == country]
    
    ax.plot(country_data['Year'], country_data['LE'], label='Life Expectancy', color='blue')
    ax.set_ylabel('Life Expectancy', color='blue')
    ax.tick_params(axis='y', labelcolor='blue')
    ax.set_xlabel('Year')
    
    ax_twin = ax.twinx()
    ax_twin.plot(country_data['Year'], country_data['GDP'], label='GDP', color='#FF00'+str(56))
    ax_twin.set_ylabel('GDP', color='red')
    ax_twin.tick_params(axis='y', labelcolor='red')
    
    ax.set_title(country)
    ax.legend(loc='upper left')
    ax_twin.legend(loc='upper right')



plt.tight_layout()
plt.show()
```
